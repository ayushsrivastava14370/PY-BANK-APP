```python
import json
import time
from getpass import getpass
import os
c=0
def main():
    global c
    if c==0:
        print('welcome to our application'.upper().center(100,' '))
        msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tApplication Start"
        fp = open('log.txt','a')
        fp.write(msg)
        fp.close()
    c+=1
    s = """1.Signup\n2.Login\n3.Exit"""
    print(s)
    ch=int(input('Enter choice from above: '))
    if ch==1:
        signup()
    elif ch==2:
        login()
    else:
        exit()
        

def signup():
    msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tRequest for Signup"
    fp = open('log.txt','a')
    fp.write(msg)
    fp.close()
    print('this is signup page')
    d = {'Name':input('Enter Name'),'Bal':int(input('Enter Balance')),'Pwd':getpass('Enter Password')}
    last_acc = int(os.listdir('bank/')[-1].split('.')[0])
    acc = last_acc+1
    f = open(f'Bank/{acc}.json','w')
    json.dump(d,f)
    f.close()
    print('Signup Successfull')
    print('Now you can login')
    print('Account Details: ')
    print('Account Number: ',acc)
    for key,value in d.items():
        print(key,':',value)
    msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tAccount Create for {acc}"
    fp = open('log.txt','a')
    fp.write(msg)
    fp.close()
    main()
    
def login():
    print('This is login page')
    acc = input('Enter account number: ') #1001.json
    path = acc + '.json'
    if path in os.listdir('bank/'):
        
        pwd = getpass('Enter Password')
        f = open(f'bank/{path}')
        data = json.load(f)
        if data['Pwd'] == pwd:
            
            s = """\n1.Debit\n2Credit\n3.Check Balance\n4.Update Password"""
            print(s)
            ch = int(input('Enter choice from above: '))
            msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tLogin Successfull for {acc}"
            fp = open('log.txt','a')
            fp.write(msg)
            fp.close()
            if ch==1:
                debit(acc,pwd)
            elif ch==2:
                credit(acc,pwd)
            elif ch==3:
                check_bal(acc,pwd)
            elif ch==4:
                update_pwd(acc,pwd)
            else:
                main()
            
        else:
            msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tLogin Failed Due to Invalid Password for {acc}"
            fp = open('log.txt','a')
            fp.write(msg)
            fp.close()
            print('Invalid Password')
            login()
    
    else:
        print('Account Not Exist')
        msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tLogin Failed due to Incorrect Account Number"
        fp = open('log.txt','a')
        fp.write(msg)
        fp.close()
        main()
        
def exit():
    msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tApplication Closed"
    fp = open('log.txt','a')
    fp.write(msg)
    fp.close()
    print('Thanks for using our application')
    for i in '..........Exciting..........':
        print(i,end='')
        time.sleep(0.09)
    

def debit(acc,pwd):
    
    f = open(f"bank/{acc}.json")
    data = json.load(f)
    bal = int(input('Enter Balance amount which you want to withdrawl: '))
    if bal > data["Bal"]:
        msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tWithdrawl Failed for Insufficient Balance for {acc}"
        fp = open('log.txt','a')
        fp.write(msg)
        fp.close()
        print('Insufficient Balance')
        debit(acc,pwd)
    else:
        data['Bal']-=bal
        f = open(f"bank/{acc}.json",'w')
        json.dump(data,f)
        f.close()
        msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tApplication Start"
        fp = open('log.txt','a')
        fp.write(msg)
        fp.close()
        print('Withdrawl Successfull')
        main()
        
def credit(acc,pwd):
    
    f = open(f"bank/{acc}.json")
    data = json.load(f)
    bal = int(input('Enter Balance amount which you want to deposit: '))
    data['Bal']+=bal
    f = open(f"bank/{acc}.json",'w')
    json.dump(data,f)
    f.close()
    print('Deposit Successfull')
    msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tDeposit Successfull for {acc}"
    fp = open('log.txt','a')
    fp.write(msg)
    fp.close()
    main()
    
def check_bal(acc,pwd):
    f = open(f"bank/{acc}.json")
    data = json.load(f)
    print('Account Balance: ',data['Bal'])
    msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tBalance Check for {acc}"
    fp = open('log.txt','a')
    fp.write(msg)
    fp.close()
    main()
    
def update_pwd(acc,pwd):
    
    pwd1 = getpass('Enter new Password')
    pwd2 = input('Enter Confirm Password')
    if pwd1==pwd2:
        f = open(f"bank/{acc}.json")
        data = json.load(f)
        data['Pwd'] = pwd1
        f = open(f"Bank/{acc}.json",'w')
        json.dump(data,f)
        f.close()
        print('Password Updated Successfully')
        msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tPassword update for {acc}"
        fp = open('log.txt','a')
        fp.write(msg)
        fp.close()
        main()
    else:
        print('Confirm Password not Matched')
        msg = f"\n{time.strftime('%d-%m-%Y')}\t\t{time.strftime('%H:%M:%S')}\t\tPassword Update Failed due to Incorrect Password for {acc}"
        fp = open('log.txt','a')
        fp.write(msg)
        fp.close()
        update_pwd(acc,pwd)
```
